```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Employee-turnover.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Attrition (Yes/No)']=df['Attrition (Yes/No)'].apply(lambda x:1 if x=='Yes' else 0)
df['Attrition (Yes/No)'].value_counts()
```

```python
plt.figure(figsize=(22,20))
sns.boxplot(data=df)
```

```python
df.columns
```

```python
q1=df['Monthly Income (USD)'].quantile(0.25)
q3=df['Monthly Income (USD)'].quantile(0.75)
iqr=q3-q1
ul=q3+1.5*iqr
ll=q1-1.5*iqr
df=df[~((df['Monthly Income (USD)']>ul)|(df['Monthly Income (USD)']<ll))]
df.shape
```

```python
yes=df[df['Attrition (Yes/No)']==1]
no=df[df['Attrition (Yes/No)']==0]
print(yes.shape)
print(no.shape)
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(data=df)
```

```python
df.hist(bins=20,figsize=(12,10))
plt.show()
```

```python
df.groupby('Attrition (Yes/No)').mean()
```

```python
no_sample=no.sample(n=100)
new_data=pd.concat([no_sample,yes],axis=0)
new_data['Attrition (Yes/No)'].value_counts()
```

```python
sns.countplot(x=new_data['Attrition (Yes/No)'])
```

```python
df.corr()['Attrition (Yes/No)'].sort_values()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
ll=LabelEncoder()
df['Job Role']=ll.fit_transform(df['Job Role'])
df['Gender']=ll.fit_transform(df['Gender'])
df['Marital Status']=ll.fit_transform(df['Marital Status'])
df['Job Satisfaction']=ll.fit_transform(df['Job Satisfaction'])
df['Environment Satisfaction']=ll.fit_transform(df['Environment Satisfaction'])
df['Work Life Balance']=ll.fit_transform(df['Work Life Balance'])
```

```python
df.head()
```

```python
df.info()
```

```python
X=df[['Job Role','Gender', 'Marital Status', 'Distance From Home (kms)', 'Job Satisfaction', 'Monthly Income (USD)', 'Salary Hike (%)', 'Years In Current Role', 'Years Since Last Promotion', 'Environment Satisfaction', 'Work Life Balance'] ]
Y=df['Attrition (Yes/No)']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7,test_size=0.3,random_state=42)
from sklearn.preprocessing import StandardScaler 
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(X_train,Y_train)
lg_pred=lg.predict(X_test)
from sklearn import metrics
print("Accuracy score: ",metrics.accuracy_score(Y_test,lg_pred))
```

```python
print("Classification report for logistic regression")
print(metrics.classification_report(Y_test,lg_pred))
```

```python
print("confusion matrix")
conf_lg=(metrics.confusion_matrix(lg_pred,Y_test))
print(conf_lg)
sns.heatmap(conf_lg,annot=True)
plt.xlabel("Predicted Attrition")
plt.ylabel("Actual Attrition")
plt.show()
```

```python
lg_pred_prob=lg.predict_proba(X_test)[::,1]
fpr,tpr,_=metrics.roc_curve(Y_test,lg_pred_prob)
auc=metrics.roc_auc_score(Y_test,lg_pred_prob)
print("AUC",auc)
plt.plot(fpr,tpr,label='auc'+str(auc))
plt.legend(loc=4)
plt.title("ROC")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
#decision tree,knn,rf
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
dt_pred=dt.predict(X_test)
print("Accuracy score for DT model: ",metrics.accuracy_score(Y_test,dt_pred))
```

```python
print("Classification report")
print(metrics.classification_report(Y_test,dt_pred))
```

```python
print("confusion matrix")
conf_dt=metrics.confusion_matrix(dt_pred,Y_test)
print(conf_dt)
sns.heatmap(conf_dt,annot=True)
plt.xlabel("Predicted Attrition")
plt.ylabel("Actual Attrition")
plt.show()
```

```python
dt_pred_prob=dt.predict_proba(X_test)[::,1]
fpr,tpr,_=metrics.roc_curve(Y_test,dt_pred_prob)
auc=metrics.roc_auc_score(dt_pred_prob,Y_test)
print("AUC: ",auc)
plt.plot(fpr,tpr,label='auc'+str(auc))
plt.legend(loc=4)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
rf_pred=rf.predict(X_test)
print("Accuracy score for RF model: ",metrics.accuracy_score(Y_test,rf_pred))
```

```python
print("Classification report")
print(metrics.classification_report(Y_test,rf_pred))
```

```python
conf_rf=metrics.confusion_matrix(Y_test,rf_pred)
print("Confusion Matrix for RF model")
print(conf_rf)
sns.heatmap(conf_rf,annot=True)
plt.xlabel("Predicted Attrition")
plt.ylabel("Actual Attrition")
plt.show()
```

```python
rf_pred_prob=rf.predict_proba(X_test)[::,1]
fpr,tpr,_=metrics.roc_curve(Y_test,rf_pred_prob)
auc=metrics.roc_auc_score(Y_test,rf_pred_prob)
print("AUC",auc)
plt.plot(fpr,tpr,label='auc'+str(auc))
plt.legend(loc=4)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
kn.fit(X_train,Y_train)
kn_pred=kn.predict(X_test)
print("Accuracy score: ",metrics.accuracy_score(Y_test,kn_pred))
```

```python
print("Classification report")
print(metrics.classification_report(Y_test,kn_pred))
```

```python
print("Confusion Matrix")
conf_kn=metrics.confusion_matrix(Y_test,kn_pred)
print(conf_kn)
sns.heatmap(conf_kn,annot=True)
plt.xlabel("Predicted Attrition")
plt.ylabel("Actual Attrition")
plt.show()
```
